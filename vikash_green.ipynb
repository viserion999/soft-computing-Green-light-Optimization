```python
#Function to find the fitness value
def fitness_function(val):
    output_=1-2*val+val**2
    return output_
```

```python
#Iterate to call the fitness function (initialization of fitness function )
initial_fitness=[]
for i in range(10):
    initial_fitness.append(fitness_function(initial_position[i]))
print(initial_fitness)
```

```python
#Local Best Position value of each particle for first iteration is 1st current position
local_best=initial_fitness
```

```python
print("Value for initally Local Best ",local_best)
```

```python
#Global Best
#Finding the GBF value which is max of LBF
global_best=max(local_best)
print(global_best)
```

```python
#Some constraint parameter in PSO
print("Now we have some constrain parameter in PSO Algo.\n")
print("The value of intial weight(W)=0.7 , and two accleration coefficient c1=2, c2=2")
w=0.7
#C1 and C2 are acceleration coeffient
c1=2
c2=2
```

```python
v=initial_velocity
current_position= initial_position
print(v,current_position,local_best)
```

```python
#Velocity updation for iteration 2 and furthur
iteration=int(input())
print("We are going to perform the no of ",iteration," iteration to optimise it.")
for j in range(1,iteration):
    
    for a in range(0,10):
        if(fitness_function(current_position[a])>local_best[a]):
            local_best[a]=fitness_function(current_position[a])
    
    if(global_best>max(local_best)):
        global_best=max(local_best)
    
    for i in range(10):
        v[i]+=(c1*np.random.rand()*(local_best[i]-current_position[i]))+(c2*np.random.rand()*(global_best-current_position[i]))
        current_position[i]+=v[i]
        
        
    
```
